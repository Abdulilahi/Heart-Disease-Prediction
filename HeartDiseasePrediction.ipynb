```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
x=df.drop(columns=["target"])
y=df["target"]

```

```python
# from sklearn.preprocessing import StandardScaler
# st=StandardScaler()
# x=st.fit_transform(x)
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(x,y,stratify=y,random_state=2,test_size=0.2)
```

```python
print(Xtrain.shape,Xtest.shape,Ytrain.shape,Ytest.shape)
```

```python
Lr=LogisticRegression()
Dtc=DecisionTreeClassifier()
Rfc=RandomForestClassifier()
Vc=VotingClassifier(estimators=[("LogisticRegression",Lr),("DecisionTreeClassifier",Dtc)],voting="soft")
Bc=BaggingClassifier()
```

```python
# fitting the data
Lr.fit(Xtrain,Ytrain)
Dtc.fit(Xtrain,Ytrain)
Rfc.fit(Xtrain,Ytrain)
Vc.fit(Xtrain,Ytrain)
Bc.fit(Xtrain,Ytrain)
```

```python
# predicting on train data
ypred_Lr=Lr.predict(Xtrain)
ypred_Dtc=Dtc.predict(Xtrain)
ypred_Rfc=Rfc.predict(Xtrain)
ypred_Vc=Vc.predict(Xtrain)
ypred_Bc=Bc.predict(Xtrain)
```

```python
print("R2 score with different models")
print("r2_score on train data by Logistic Regression: ",accuracy_score(Ytrain,ypred_Lr))
print("r2_score on train data by DecisionTreeClassifier: ",accuracy_score(Ytrain,ypred_Dtc))
print("r2_score on train data by RandomForest Classifier: ",accuracy_score(Ytrain,ypred_Rfc))
print("r2_score on train data by Voting Classifier: ",accuracy_score(Ytrain,ypred_Vc))
print("r2_score on train data by Bagging Classifier: ",accuracy_score(Ytrain,ypred_Bc))
```

```python
# predicting on train data
ypred_Lr=Lr.predict(Xtest)
ypred_Dtc=Dtc.predict(Xtest)
ypred_Rfc=Rfc.predict(Xtest)
ypred_Vc=Vc.predict(Xtest)
ypred_Bc=Bc.predict(Xtest)
```

```python
print("R2 score with different models")
print("r2_score on test data by Logistic Regression: ",accuracy_score(Ytest,ypred_Lr))
print("r2_score on test data by DecisionTreeClassifier: ",accuracy_score(Ytest,ypred_Dtc))
print("r2_score on test data by RandomForest Classifier: ",accuracy_score(Ytest,ypred_Rfc))
print("r2_score on test data by Voting Classifier: ",accuracy_score(Ytest,ypred_Vc))
print("r2_score on test data by Bagging Classifier: ",accuracy_score(Ytest,ypred_Bc))
```
